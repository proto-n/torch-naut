```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mdn = pd.read_csv('results/scaling/mdn.csv')
crps = pd.read_csv('results/scaling/crps.csv')
```

```python
plt.figure(figsize=(4,3))
plt.title('Training 10 epochs')
plt.plot(mdn['nparams'], mdn['time'], 'x-', label='mdn')
plt.plot(crps['nparams'], crps['time'], 'x-', label='crps')
plt.xlabel('number of parameters')
plt.ylabel('runtime (s)')
plt.legend()
plt.tight_layout()
plt.savefig('plots/scaling.pdf')
```

```python
plt.figure(figsize=(4,3))
plt.title('Training 10 epochs (log-log)')
plt.plot(mdn['nparams'], mdn['time'], 'x-', label='mdn')
plt.plot(crps['nparams'], crps['time'], 'x-', label='crps')
plt.xlabel('number of parameters (log)')
plt.ylabel('runtime (s) (log)')
plt.legend()
plt.yscale('log')
plt.xscale('log')
plt.tight_layout()
plt.savefig('plots/scaling_log.pdf')
```
